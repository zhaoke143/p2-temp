```python
import pandas as pd
import operator
import numpy as py
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
df = pd.read_csv("../globalterrorismdb.csv", encoding = "latin-1", low_memory=False, 
                 usecols=['eventid', 'iyear','imonth','iday','success','nkill', 'nkillus', 'nwound', 'nwoundus','country',
                          'country_txt','region','region_txt','city','latitude','longitude',
                          'attacktype1','attacktype1_txt','targtype1','targtype1_txt','natlty1',
                          'natlty1_txt','weaptype1','weaptype1_txt', 'weapdetail'])
df.dropna(subset=['nkill'], inplace = True)
df.drop(df[df['nkill'] == 0].index , inplace=True)
df.dropna(subset=['latitude', 'longitude'], inplace = True)
df['location'] = df[['latitude', 'longitude']].values.tolist()
df.dropna(subset=['location'], inplace = True)
df ['location'] = pd.to_numeric(df['location'], errors='coerce')
#pd.DataFram.astype({'location': float})
df.head(50)
```

```python
reg = df[['region', 'region_txt']]
reg_num = dict(zip(reg.region, reg.region_txt))
reg_num
```

```python
country = df[['country', 'country_txt']]
c_name = country.loc[country['country_txt'] == 'Iraq']
c_num = dict(zip(country.country, country.country_txt))
sorted_c = sorted(c_num.items(), key=operator.itemgetter(0))
sorted_c
```
